```python
# DataFrame.loc 
# 是 Pandas 中用于基于标签的索引方式，它后面通常跟行标签和列标签作为参数。
# 行标签通常是索引（index），而列标签是列名（column names）。

import pandas as pd

# 创建一个简单的DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}, index=['x', 'y', 'z'])

# 使用.loc选择单个元素
# 选择行标签为'x'且列标签为'A'的元素
element = df.loc['x', 'A']
print(element)  # 输出: 1

# 使用.loc选择多行多列
# 选择行标签为'x'和'y'的所有列
subset = df.loc[['x', 'y']]
print(subset)

# 选择行标签为'x'和'y'，列标签为'A'和'B'的子集
subset = df.loc[['x', 'y'], ['A', 'B']]
print(subset)

# 使用.loc选择单列
# 选择列标签为'A'的所有行
column_subset = df.loc[:, 'A']
print(column_subset)

# 使用.loc选择单行
# 选择行标签为'y'的所有列
row_subset = df.loc['y']
print(row_subset)

# 使用.loc选择满足条件的行
# 选择列'A'值大于1的所有行
condition_subset = df.loc[df['A'] > 1]
print(condition_subset)
```

```python
## 筛选完以后，使用loc, 如何输出其他列
is_long = movies.duration >= 200
# movies.loc[movies.duration >= 200, 'genre']
movies.loc[is_long, 'genre']
```

```python
## 如果多个列有筛选条件 
# use **`&`** instead of **`and`** 
# use **`|`** instead of **`or`**)
movies[(movies.duration >= 200)&(movies.genre=='Drama')]
```

```python
## 如果对一列里的行有筛选条件
## 针对object
movies[(movies.genre=='Drama')|(movies.genre=='Crime')|(movies.genre=='Action')]
## 或
movies[movies.genre.isin(['Drama','Crime','Action'])].head(10)
```

```python
## 如果对一列里的行有筛选条件
## 针对float64
movies[(movies.duration>=200)&(movies.duration<=220)&(movies.genre.isin(['Drama','Crime','Action']))]
```

```python
## 模糊查询
movies[movies.title.str.contains("Part II", case=False,  regex=True)]
```
