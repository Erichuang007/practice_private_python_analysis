```python
import pandas as pd  
  
# 假设你有一个名为 df 的 DataFrame  
df = pd.DataFrame({  
    'A': [1, 1, 2, 2, 3, 3],  
    'B': [1, 1, 1, 2, 2, 2],  
    'C': [4, 3, 2, 1, 5, 6]  
})  
  
# 首先，按照 A 和 B 列进行分组，并按照 C 列排序  
df_sorted = df.sort_values(by=['A', 'B', 'C'])  
  
# 接下来，在排序后的 DataFrame 上使用 rank 方法，但注意 rank 方法默认对整个 DataFrame 进行排名  
# 我们需要分组后对每个组内的数据单独排名  
# 首先创建一个用于分组的键  
grouped = df_sorted.groupby(['A', 'B'])  
  
# 定义一个辅助函数，对每个组应用 rank 方法  
def rank_within_group(group):  
    group['Rank'] = group['C'].rank(method='min', ascending=True)  
    return group  
  
# 应用辅助函数到每个分组  
df_with_rank = grouped.apply(rank_within_group)  
  
# 显示结果  
df_with_rank
```

```python
# DataFrame.loc 
# 是 Pandas 中用于基于标签的索引方式，它后面通常跟行标签和列标签作为参数。
# 行标签通常是索引（index），而列标签是列名（column names）。

import pandas as pd

# 创建一个简单的DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}, index=['x', 'y', 'z'])

# 使用.loc选择单个元素
# 选择行标签为'x'且列标签为'A'的元素
element = df.loc['x', 'A']
print(element)  # 输出: 1

# 使用.loc选择多行多列
# 选择行标签为'x'和'y'的所有列
subset = df.loc[['x', 'y']]
print(subset)

# 选择行标签为'x'和'y'，列标签为'A'和'B'的子集
subset = df.loc[['x', 'y'], ['A', 'B']]
print(subset)

# 使用.loc选择单列
# 选择列标签为'A'的所有行
column_subset = df.loc[:, 'A']
print(column_subset)

# 使用.loc选择单行
# 选择行标签为'y'的所有列
row_subset = df.loc['y']
print(row_subset)

# 使用.loc选择满足条件的行
# 选择列'A'值大于1的所有行
condition_subset = df.loc[df['A'] > 1]
print(condition_subset)
```
